# 데이터 전처리

```python
# 데이터 확인
df = pd.read_excel('./total_data.xlsx',  names = ['date', 'ultrafine dust', 'fine dust', 'SO2', 'O3', 'NO2', 'CO', 'area'])

# '점검중'을 0으로, '-'를 NAN 으로 변경
df = df.replace('점검중', 0)
df = df.replace('-', np.NaN)

df.head()
```

```python
# 새로운 빈 데이터프레임 생성
new_df = pd.DataFrame()
new_df
```

```python
# 월, 일 을 분리하여 new_df에 컬럼 추가
new_df['month'] = df.date.str.split('월').str[0]
new_df['day'] = df.date.str.split('월').str[1].str[:3]
new_df
```

```python
# df['new']= df['month'].apply(lambda x: 10 if x >5 )
```

```python
new_df['ultrafine dust'] = df['ultrafine dust']
new_df['fine dust'] = df['fine dust']
new_df['SO2'] = df['SO2']
new_df['O3'] = df['O3']
new_df['NO2'] = df['NO2']
new_df['CO'] = df['CO']
new_df['area'] = df['area']

# 온천장, 초량동에서 '(도로변)' 삭제
new_df = new_df.replace('온천동(도로변)', '온천동')
new_df = new_df.replace('초량동(도로변)', '초량동')

# sum(new_df['area'] == '초량동')
```

```python
# 지역별, 년별 평균
areayear = new_df.groupby('area').mean()

# 인덱스로 지정되어 있는 'area'를 컬럼으로 빼주기
areayear.reset_index(inplace=True)

# len(areayear) -> 33
```

```python
areayear
```

```python
areayear['area'] 
```

```python
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
# 지역별, 년별 평균 막대그래프

x = np.arange(33)
area = areayear['area'] 
ultrafine_dust = areayear['ultrafine dust']

plt.figure(figsize = (13, 8))

# 그래프 바 색상 변경
colors = sns.color_palette('hls', len(area))

plt.bar(x, ultrafine_dust, color = colors)
plt.xticks(x, area, fontsize=10, rotation=40)

plt.show()
```

```python
# 지역별, 월별 평균
areamonth = new_df.groupby(['area', 'month']).mean()

# 인덱스로 지정되어 있는 'area', 'month'를 컬럼으로 빼주기
#areamonth.reset_index(inplace=True)

areamonth
```

```python
gaeguem = areamonth.loc[['개금동']]
gaeguem.reset_index('month', inplace=True)
gaeguem
```

```python
# 지역별, 월별 평균 막대 그래프(ultrafine dust만)


# 개금동
gaeguem = areamonth.loc[['개금동']]
gaeguem.reset_index('month', inplace=True)

x = np.arange(9)
month = gaeguem['month']
ultrafine_dust = gaeguem['ultrafine dust']

plt.figure(figsize = (13, 8))

# 그래프 바 색상 변경
colors = sns.color_palette('hls', len(month))

plt.bar(x, ultrafine_dust, color = colors)
plt.xticks(x, month+'월', fontsize=15)

plt.title('개금동 월별 데이터', fontsize=20)
plt.show()


# 광복동
gwangbog = areamonth.loc[['광복동']]
gwangbog.reset_index('month', inplace=True)

x = np.arange(9)
month = gwangbog['month']
ultrafine_dust = gwangbog['ultrafine dust']

plt.figure(figsize = (13, 8))

colors = sns.color_palette('hls', len(month))

plt.bar(x, ultrafine_dust, color = colors)
plt.xticks(x, month+'월', fontsize=15)

plt.title('광복동 월별 데이터', fontsize=20)
plt.show()
```

```python
# 인덱스로 지정되어 있는 'area', 'month'를 컬럼으로 빼주기
# (반복 실행되면 안됨)

gaeguem.reset_index(inplace=True)
```

```python
gaeguem.iloc[:, :3]
```

```python
# MinMaxScaler(0~1 스케일링)

from sklearn.preprocessing import MinMaxScaler

# 데이터프레임 분할 -> area, month / 그 외
test_gaeguem = gaeguem.iloc[:, 2:]

x = test_gaeguem.values
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
scaled_gaeguem = pd.DataFrame(x_scaled)
scaled_gaeguem.columns = ['ultrafine dust', 'fine dust', 'SO2', 'O3', 'NO2', 'CO']

# 데이터프레임 합치기
scaled_gaeguem = pd.concat([gaeguem.iloc[:, :2], scaled_gaeguem], axis = 1)
scaled_gaeguem
```

```python
# 지역별, 월별 평균 막대 그래프(전체)


# 개금동
x = np.arange(9)
month = scaled_gaeguem['month']
ultrafine_dust = scaled_gaeguem['ultrafine dust']
fine_dust = scaled_gaeguem['fine dust']
SO2 = scaled_gaeguem['SO2']
O3 = scaled_gaeguem['O3']
NO2 = scaled_gaeguem['NO2']
CO = scaled_gaeguem['CO']
    
plt.figure(figsize = (13, 8))

# 그래프 바 색상 변경
colors = sns.color_palette('hls', len(month))

plt.plot(x, ultrafine_dust, label = 'ultrafine_dust')
plt.plot(x, fine_dust, label = 'fine_dust')
plt.plot(x, SO2, label = 'SO2')
plt.plot(x, O3, label = 'O3')
plt.plot(x, NO2, label = 'NO2')
plt.plot(x, CO, label = 'CO')

plt.xticks(x, month+'월', fontsize=15)

plt.title('개금동 월별 데이터', fontsize=20)
plt.legend()
plt.show()
```
